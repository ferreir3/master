Now lets repeat this process for all the neighbouhoods of our list. For this porpouse, I will define the following function:

```python
def getVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

### CHECK
    venuesList = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    venuesList.columns = ['Neighbourward', 
                  'Neighbourward Latitude', 
                  'Neighbourward Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(venuesList)
```

```python
number_mostFrequeest_venues = 5
for ward in groupedVenues['Neighbourward']:
    print("===-"+ward+"===")
    temp = groupedVenues[groupedVenues['Neighbourward'] == ward].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float) # converting values to float
    temp = temp.round({'freq': 2}) # rounding to 2 decimal digits
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(number_mostFrequeest_venues)) #printing values with the fre  uency in ascing order
    print('\n')
```

Let's integrate this data into a pandas dataframe. For this purpose, the following function sorts the venues in descending order.

```python
def return_most_common_venues(row, number_mostFrequeest_venues):
    rowCategorieType = row.iloc[1:] # we are dropping the neighbouward name and only are selecting the venue fecuency
    rowCategorieType_sorted = rowCategorieType.sort_values(ascending=False)
    
    return rowCategorieType_sorted.index.values[0:number_mostFrequeest_venues]
```

Now let's define a new dataframe and display the most common 10 venues for each neighborward.

```python
number_mostFrequeest_venues = 10

indicators = ['st', 'nd', 'rd']

# we need tocreate columns to fit the number of top venues
import numpy as np
columns = ['Neighbourward']
for ind in np.arange(number_mostFrequeest_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))
        
# and now is when I define a new dataframe
wards_venues_sorted = pd.DataFrame(columns=columns)
```
