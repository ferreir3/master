```python
#lets load relevant libraries
!conda install -c anaconda beautifulsoup4 #this works
#!conda install -c conda-forge beautifulsoup4
#install -c conda beautifulsoup4
from bs4 import  BeautifulSoup
import requests
import pandas as pd
```

lets find the table

```python
# lets define the url address for extracting information
wiki_url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
#lets download the page
response = requests.get(wiki_url)
```

```python
# lets parse the dictionary with beautifull soup for extracting the postal codes and their respective names
soup1 = BeautifulSoup(response.text,'html.parser')
```

```python
table_toronto = soup1.find_all("tbody")[0]
```

```python
neighBourwardCode = []
boroughList = []
neighBourwardList = []
#new_table = pd.DataFrame(columns=range(162,2), index = [0]) # I know the size 
row_marker = 0
for row in table_toronto.find_all('tr'):
    print("==============")
    #print(row)
    column_marker = 0
    columns = row.find_all('td')
    for column in columns:
        print("..................")
        #print(column.get_text())
        if(column.get_text()[4:7]=='Not'):
            continue
        else:
            neighBourwardCode.append(column.get_text()[1:4])
            text = column.get_text()[4:-2]
            for index,values in enumerate(text):
                if(text[index+1]=='('):
                    boroughList.append(column.get_text()[4:index+5]) # to avoid "\n" characters
                    neighBourwardList.append(column.get_text()[index+6:-2]) # to avoid ")\n" characters
                    if(neighBourwardList[-1][-1]==')'):
                        neighBourwardList[-1] = neighBourwardList[-1][:-1]
                    neighBourwardList[-1] = neighBourwardList[-1].replace('(',' ')
                    neighBourwardList[-1] = neighBourwardList[-1].replace(')','')
                    neighBourwardList[-1] = neighBourwardList[-1].replace(' / ',', ')
                    #for index2, value2 in enumerate(neighBourwardList[-1]):
                    break
                else:
                    continue
            print(neighBourwardCode[-1])
            print(boroughList[-1])
            print(neighBourwardList[-1])
            #new_table.iat[row_marker,column_marker] = column.get_text()
            column_marker += 1
neighBourwardDict = {'Postal Code':neighBourwardCode,'Borough':boroughList,'Neighbourward':neighBourwardList}
```

Using the dictionary "neighBourwardDict" lets create a data frame

```python
dataF1 = pd.DataFrame(neighBourwardDict)
```
